# Metropolis-Hastings vs EMCEE

As you may know the Metropolis-Hastings algorithm works by proposing an initial position on the configuration space 
and then randomly moving to another place (random walk). We decide to stay on this spot and save it only if the probability of the new step is 
larger than the previous step, if not we stay in the same spot and take a new random step.

The affine-invariant ensemble sampler consist on generating an enable of “walkers” where the movement of each walker depends on the complementary ensemble composed of all the remaining walkers. To move the walkers to a new position on the parameter space we follow the same principle as the M-H algorithm but with a different random variable to obtain the new random position. What EMCEE does is that it parallelize the sampler to run all the walker simultaneously by splitting the ensemble into two sub-ensembles and repeating  the previous algorithm comparing the probabilities of the sub-ensembles.

The usefulness of this sampler is that it can outperform a M-H algorithm on convergence speed, meaning that we need fewer number of steps to find a high probability density zone and because we can initialize every walker in different positions of the parameter space we could sample multimodal distributions better (there are other samplers best suited for that task). For a more detailed description of the workings of emcee check this couple of papers:

https://arxiv.org/pdf/1202.3665.pdf

https://msp.org/camcos/2010/5-1/camcos-v5-n1-p04-s.pdf

In this case the $\mu$ is the distance modulus of the supernova difined as 

$\mu = 25 - 5log_{10}h+5log_{10}(D^*_L)$

where $D^*_L = D_L(h=1)$ is the luminosity distance, that for a flat universe can be defined as:

$D_L(z) = \frac{c}{H0}(1+z)\left[\eta(1,\Omega_m)-\eta(\frac{1}{1+z},\Omega_m)\right]$,

with:

$\eta(a,\Omega_m) = 2\sqrt{s^3+1}\left[\frac{1}{a^4}-0.1540\frac{s}{a^3} + 0.4304\frac{s^2}{a^2} + 0.19097\frac{s^3}{a} + 0.066941s^4\right]^{-1/8}$

and: $s^3=(1-\Omega_m)/\Omega_m$

We start by construncting functions for $D_L(z)$, $\eta$ and $\mu$

```python
def eta(pars,a):
    Omega_m = pars
    s = ((1.0-Omega_m)/Omega_m)**(1./3.)
    
    coefs = [0.1540,0.4304,0.19097,0.066941]
    
    val = a**-4 - coefs[0]*(s/a**3) + coefs[1]*(s/a)**2 + coefs[2]*(s**3/a) + coefs[3]*(s**4)
    
    return 2.0*np.sqrt(s**3 + 1)*(val**(-1./8.))
```

```python
def D_L(pars,z):
    Omega_m = pars
    a_0 = 1.0
    a = 1./(1. + z)
    c = 299792.0 #km/s
    
    val = eta(Omega_m,a_0) - eta(Omega_m,a)
    
    return (c/(a*100.)) * val
```

```python
def mu(pars,z):
    Omega_m,h = pars
    
    val = 25.0 - (5.0 * np.log10(h)) + (5.0 * np.log10(D_L(Omega_m,z)))
    
    return val
```

# Now we define our Likelihood and priors 

We use the General Gaussian Likelihhod

$\text{Log}(\text{L}) \approx -\frac{1}{2}\left([D-M]^T C^{-1} [D-M]\right) $

```python
def log_like(pars,z,data,cov):
    model = mu(pars,z)
    
    coefs = data - model
    
    cov_inv = np.linalg.inv(cov)
    p = np.dot(coefs.T,np.dot(cov_inv,coefs))
    return -0.5*p
```

```python
def priors(pars,plist):
    Omega_m, h = pars
    
    if plist[0]<Omega_m<plist[1] and plist[2]<h<plist[3]:
        return 0.0
    return -np.inf
```

```python
def log_post(pars,z,data,cov,plist):
    prior = priors(pars,plist)
    if not np.isfinite(prior):
        return -np.inf
    return prior + log_like(pars,z,data,cov)
```

# Now the complete model

As i mentioned before, the expresion for $D_L(z)$ is an approximation that is only valid for a flat universe $\Omega = \Omega_m + \Omega_v = 1$. The more general way to calculate the Luminosity Distance is 

$D_L(z) = \frac{(1+z)c}{H_0\sqrt{|1-\Omega|}}S_k(r)$

where $S_k(r) = \text{sin}r, r, \text{sinh}r$ if $\Omega >1, =1$ or $<1$ and $r(z)$ 

$r(z) = \sqrt{|1-\Omega|}\int_0^z\frac{dz'}{\sqrt{\Omega_m(1+z')^3+\Omega_v+(1-\Omega)(1+z')^2}}$

For more information read: https://www.imperial.ac.uk/media/imperial-college/research-centres-and-groups/astrophysics/public/icic/data-analysis-workshop/2016/Preliminary-Exercise-2016_0.pdf
